Costruzione dello stato

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
import numpy as np

alpha = np.sqrt(0.5)
beta = np.sqrt(0.5)

qc = QuantumCircuit(3,2)    # 3 qubit e 1 bit classico

#inizializza il primo qubit
qc.initialize([alpha, beta], 0)

#entangla i qubit ancilla
qc.cx(0, 1)
qc.cx(0, 2)

qc.draw("mpl")
```

Implementazione la trasformazione unitaria che simula l'errore

```python
#applica una piccola rotazione Y al secondo qubit
qc.ry(np.pi / 10, 1)

qc.draw("mpl")
```

Implementazione del codice di correzione degli errori

```python
#misura i qubit ancilla
qc.measure([1,2], [0,1])

qc.draw("mpl")
```

Esegui il circuito quantistico

```python
#transpila il circuito
qc = transpile(qc, Aer.get_backend('qasm_simulator'))

#esegui il circuito
backend = Aer.get_backend('qasm_simulator')
result = backend.run(qc).result()
counts = result.get_counts(qc)

#se i risultati un bit flip, applica una correzione
if '01' in counts:
    qc.x(1)
elif '10' in counts:
    qc.x(2)
elif '11' in counts:
    qc.x(0)

#misura il primo qubit
qc.measure(0, 0)

#transpila il circuito
qc = transpile(qc, Aer.get_backend('qasm_simulator'))

#esegui il circuito
result = backend.run(qc).result()

print(result.get_counts(qc))
```

Disegna il circuito finale

```python
qc.draw("mpl")
```
